```python
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import random
import cv2
```

Загрузим исходное изображение и покажем его:

```python
img = image.load_img("../input_image.jpg")
plt.imshow(img)
```

Преобразуем исходное изображение в матрицу с разменостью (x, y, 3):

```python
image_matrix = image.img_to_array(img)
```

Данная функция работает крайне просто: достаточно уменьшит изображение в несколько раз(переенная factor), а после вернуть его в исходное разрешение.  При этом используется линейная интерполяция

```python
def degrade_image(img, factor):               
    h, w, _ = img.shape
    new_height = int(h / factor)
    new_width = int(w / factor)
    img = cv2.resize(img, (new_width, new_height), interpolation = cv2.INTER_LINEAR)
    img = cv2.resize(img, (w, h), interpolation = cv2.INTER_LINEAR)
    return img
```

Применим вышеописанную функцию:

```python
output_img_matrix = degrade_image(image_matrix, 4)
```

Преобразуем полученную матрицу в изображение и покажем её:

```python
output_image = image.array_to_img(output_img_matrix)
plt.imshow(output_image)
```

Сохраним полученное изображение:

```python
output_image.save("degradate_augmentation_output.jpg")
```
